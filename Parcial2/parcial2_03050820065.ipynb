```python
import numpy as np
import pandas as pd 
import time as time
```

```python
#Generar una estructura de datos que almacene 10 millones de puntos en una distribuci√≥n normal con una media de 500 y una escala de 30

datos = list(np.random.normal(500, 30, 10000000))
print(datos)
```

```python
#Se suman los valores menores a 500,000
inicio = time.time()
menor500 = []
for dato in datos:
    if dato < 500000:
        menor500.append(dato)
print(sum(menor500))       
print('Duracion: {} segundos'.format(time.time() - inicio))
```

```python
#Se cambio la lista a un dataframe de pandas para optimizar la suma
inicio = time.time()
menor500 = pd.DataFrame(datos) 
print(sum(menor500))
print('Duracion: {} segundos'.format(time.time() - inicio))
```
