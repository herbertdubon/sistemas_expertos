```python
import numpy as np
import pandas as pd 
import time as time
import matplotlib.pyplot as plt
```

```python
#Generar una estructura de datos que almacene 10 millones de puntos en una distribucin normal con una media de 500 y una escala de 30

datos = np.random.normal(500, 30, 10000000)
#print(datos)
plt.hist(datos, 1000)
plt.show()
```

```python
#Se suman los valores menores a 500,000
inicio = time.time()
menor500K = []
for dato in datos:
    if dato < 500000:
        menor500K.append(dato)
print("La cantidad total de elementos menores a 500,000 es: {}".format(len(menor500K))) 
print("La suma total de los valores menores a 500,000 es: {}".format(sum(menor500K)))  
print('Duracion: {} segundos'.format(time.time() - inicio))
```

```python
#Se cambio la lista a un dataframe de pandas para optimizar la suma
inicio = time.time()
df = pd.DataFrame(datos) 
menor500K = (datos[datos <= 500000])
print("La cantidad total de elementos menores a 500,000 es: {}".format(len(menor500K))) 
print("La suma total de los valores menores a 500,000 es: {}".format(sum(menor500K)))  
print('Duracion: {} segundos'.format(time.time() - inicio))
```
